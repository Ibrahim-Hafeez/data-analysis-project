```python
import pandas as pd
df_audit = pd.read_csv("/content/sample_data/california_housing_train.csv")

df_audit.head()
```

```python
r = df_audit['total_rooms'].corr(df_audit['total_bedrooms'])
print("Pearson Correlation (r):", round(r, 4))
```

```python
r = df_audit['housing_median_age'].corr(df_audit['median_income'])
r_squared = r ** 2
print("r^2 value:", r_squared)
print("Percentage:", round(r_squared * 100, 2),"%")
```

```python
pearson = df_audit['population'].corr(df_audit['median_house_value'], method='pearson')
spearman = df_audit['population'].corr(df_audit['median_house_value'], method='spearman')

print("Pearson (r):", round(pearson, 4))
print("Spearman (r):", round(spearman,4))
```

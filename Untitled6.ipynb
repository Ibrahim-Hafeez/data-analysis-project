```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/CodeVinayak/Student-Performance-Analysis-Machine-Learning/master/StudentsPerformance.csv'

df = pd.read_csv(url)
```

```python
df.head()
```

```python
scores_subset = df[['gender', 'math_score', 'reading_score']]
```

```python
scores_subset.head()
```

```python
middle_sample = df.iloc[10:20, df.columns.get_loc('math_score')]

middle_sample
```

```python
high_math_students = df.loc[df['math_score'] > 50, ['gender', 'writing_score']]

high_math_students
```

```python
literacy_excellence = df[
    (df['reading_score'] > 75) &
    (df['writing_score'] > 70)
]

literacy_excellence.head()
```
